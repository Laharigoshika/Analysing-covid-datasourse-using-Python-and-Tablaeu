```python
import pandas as pd
df4 = pd.read_excel(r'/Users/laharigoshika/Downloads/Simplified Car Sales Data.xlsx')
df4
#total sales and total costs for the Germany
Germany= df4[df4['CountryName']=='Germany']
Germany
Germany_totals=Germany.sum(axis=0, numeric_only=True)
Germany_total_saleprice= Germany_totals['SalePrice']
Germany_total_costprice= Germany_totals['CostPrice']
print(Germany_total_saleprice)
print(Germany_total_costprice)

#total sales and total costs for the Spain
Spain= df4[df4['CountryName']=='Spain']
Spain_totals=Spain.sum(axis=0, numeric_only=True)
Spain_total_saleprice= Spain_totals['SalePrice']
Spain_total_costprice= Spain_totals['CostPrice']
print(Spain_total_saleprice)
print(Spain_total_costprice)

#total sales and total costs for the Switzerland
Switzerland= df4[df4['CountryName']=='Switzerland']
Switzerland_totals=Switzerland.sum(axis=0, numeric_only=True)
Switzerland_total_saleprice= Switzerland_totals['SalePrice']
Switzerland_total_costprice= Switzerland_totals['CostPrice']
print(Switzerland_total_saleprice)
print(Switzerland_total_costprice)

totals= pd.DataFrame([('Germany', Germany_total_saleprice, Germany_total_costprice),
                     ('Spain', Spain_total_saleprice, Spain_total_costprice),
                     ('Switzerland', Switzerland_total_saleprice, Switzerland_total_costprice)],
                    columns= ['Country Name','Total Sales price', 'Total Cost Price'])
totals
totals.plot('Country Name',['Total Sales price','Total Cost Price'], kind="bar")
totals.plot('Country Name',['Total Sales price','Total Cost Price'], kind="barh")
```
